```python
from OCC.Core.gp import gp_Pnt, gp_Lin, gp_Dir, gp_Pln, gp_Ax3
from OCC.Core.BRepBuilderAPI import BRepBuilderAPI_MakeEdge, BRepBuilderAPI_MakeFace
from OCC.Core.BRepAlgoAPI import BRepAlgoAPI_Section
from OCC.Display.SimpleGui import init_display
from OCC.Core.TopExp import TopExp_Explorer
from OCC.Core.TopAbs import TopAbs_VERTEX
from OCC.Core.BRep import BRep_Tool

# 初始化顯示窗口
display, start_display, add_menu, add_function_to_menu = init_display()

# 創建一條線段
line_start = gp_Pnt(0, 0, 0)
line_end = gp_Pnt(10, 10, 10)
line = BRepBuilderAPI_MakeEdge(line_start, line_end).Edge()

# 創建多個平面
planes = []
for i in range(5):
    plane_origin = gp_Pnt(0, 0, i * 2)
    plane_normal = gp_Dir(0, 0, 1)
    plane = gp_Pln(plane_origin, plane_normal)
    face = BRepBuilderAPI_MakeFace(plane).Face()
    planes.append(face)

# 計算交點
intersection_points = []
for plane in planes:
    section = BRepAlgoAPI_Section(line, plane)
    section.Build()
    if section.IsDone():
        ex = TopExp_Explorer(section.Shape(), TopAbs_VERTEX)
        while ex.More():
            vertex = ex.Current()
            point = BRep_Tool.Pnt(vertex)
            intersection_points.append(point)
            ex.Next()

# 打印交點數量和座標
print(f"Number of intersection points: {len(intersection_points)}")
for point in intersection_points:
    print(f"Intersection point: ({point.X()}, {point.Y()}, {point.Z()})")

# 可視化線段和平面
display.DisplayShape(line, update=True)
for plane in planes:
    display.DisplayShape(plane, update=True)
for point in intersection_points:
    display.DisplayShape(point, update=True)

# 開始顯示
start_display()
```

```python
from OCC.Core.IGESControl import IGESControl_Reader
from OCC.Core.gp import gp_Pnt
from OCC.Core.BRepBuilderAPI import BRepBuilderAPI_MakeEdge
from OCC.Core.BRepAlgoAPI import BRepAlgoAPI_Section
from OCC.Core.TopExp import TopExp_Explorer
from OCC.Core.TopAbs import TopAbs_VERTEX
from OCC.Core.BRep import BRep_Tool
from OCC.Display.SimpleGui import init_display

# 初始化顯示窗口
display, start_display, add_menu, add_function_to_menu = init_display()

# 讀取IGS文件
def read_igs_file(file_path):
    iges_reader = IGESControl_Reader()
    iges_reader.ReadFile(file_path)
    iges_reader.TransferRoots()
    shape = iges_reader.OneShape()
    return shape

# 設定IGS文件路徑
file_path = 'path/to/your/model.igs'  # 替換為你的IGS文件路徑

# 讀取模型
model_shape = read_igs_file(file_path)

# 在模型中顯示
display.DisplayShape(model_shape, update=True)

# 創建一條線段
line_start = gp_Pnt(0, 0, 0)  # 起點
line_end = gp_Pnt(100, 100, 100)  # 終點，根據需要調整
line = BRepBuilderAPI_MakeEdge(line_start, line_end).Edge()

# 計算交點
section = BRepAlgoAPI_Section(line, model_shape)
section.Build()

# 提取交點
intersection_points = []
if section.IsDone():
    ex = TopExp_Explorer(section.Shape(), TopAbs_VERTEX)
    while ex.More():
        vertex = ex.Current()
        point = BRep_Tool.Pnt(vertex)
        intersection_points.append(point)
        ex.Next()

# 打印交點數量和座標
print(f"Number of intersection points: {len(intersection_points)}")
for point in intersection_points:
    print(f"Intersection point: ({point.X()}, {point.Y()}, {point.Z()})")

# 可視化線段和交點
display.DisplayShape(line, update=True)
for point in intersection_points:
    display.DisplayShape(point, update=True)

# 開始顯示
start_display()
```
